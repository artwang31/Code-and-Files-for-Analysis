```python
def listToString(s):
    str1 = ""
    for hey in s:
        str1 += str(hey) + ','
    return str1[:-1]
```

```python
listToString('artemas')
```

```python
def transform_to_text_list(listed_names):
    list_len = len(listed_names.split(','))
    i = 0
    output = "("
    while i < list_len:
        output = output + "'" + listed_names.split(',')[i].strip() + "'"
        if i < list_len-1:
                output = output + ','
        i += 1
    output = output + ")"
    return output
```

```python
class ArtemasDog(): # every new word should have a capital
    animal = 'dog'
    def __init__(self, breed, name, spots): #init is constructor for a class, when you create an object, it is expecting you to create a parameter breed
        #Attributes
        self.breed = breed #characteristics of the object
        self.name = name
        self.spots = spots
```
